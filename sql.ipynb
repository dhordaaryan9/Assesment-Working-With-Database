```python
# 1. Develop a Python-based ETL script using SQLAlchemy to automate the creation of the analytics_db schema and enforce composite primary keys on the ratings table.


from sqlalchemy import create_engine, Column, Integer, String, Float, PrimaryKeyConstraint
from sqlalchemy.orm import declarative_base

Base = declarative_base()

class Rating(Base):
    __tablename__ = 'ratings'
    
    restaurant_id = Column(Integer)
    user_id = Column(Integer)
    rating_score = Column(Float)
    review_text = Column(String(500))

    __table_args__ = (
        PrimaryKeyConstraint('restaurant_id', 'user_id', name='composite_pk_rating'),
    )

def setup_database():
    database_url = "mysql+pymysql://root:password@localhost/analytics_db"
    
    print("Connecting to the database...")
    engine = create_engine(database_url)
    
    print("Creating the schema and enforcing composite primary keys...")
    Base.metadata.create_all(engine)
    
    print("analytics_db schema created successfully!")

if __name__ == "__main__":
    setup_database()
```

```python
# 3. Write a Python validation script using Pandas to clean the Zomato dataset,handling "NEW" or "-" values in the rating column before relational ingestion.

import pandas as pd
import numpy as np

def validate_zomato_ratings(file_path):
    print(f"Loading data from {file_path}...")
    df = pd.read_csv(file_path)

    print("Starting the cleaning process for the 'rate' column...")

    def clean_rate(value):
        if value == 'NEW' or value == '-':
            return np.nan
        
        if isinstance(value, str):
            value = value.split('/')[0].strip()
            return float(value)
        
        return value

    df['rate'] = df['rate'].apply(clean_rate)

    null_count = df['rate'].isnull().sum()
    total_rows = len(df)
    
    print("-" * 30)
    print(f"Validation Report:")
    print(f"Total Rows Processed: {total_rows}")
    print(f"Rows with 'NEW' or '-' (now Null): {null_count}")
    print(f"Average Restaurant Rating: {df['rate'].mean():.2f}")
    print("-" * 30)

    df['rate'] = df['rate'].fillna(0)

    return df

# Example Usage:
# cleaned_df = validate_zomato_ratings('zomato.csv')
# cleaned_df.to_csv('zomato_cleaned.csv', index=False)
```
